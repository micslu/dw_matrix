```
df = pd.read_csv('data/shoes_prices.csv', low_memory=False)
df.shape
```

```
df.columns
```

```
mean_price = np.mean(df['prices_amountmin'])
mean_price
```

```
y_true = df['prices_amountmin']
price_log_mean = np.expm1(np.mean(np.log1p(y_true)))
y_pred = [np.median(price_log_mean)] * y_true.shape[0]
```

```
mean_absolute_error(y_true, y_pred)
```

```
df['brand_cat'] = df['brand'].factorize()[0]
```

```
def run_model(feats):
  X = df[ feats ].values
  y = df['prices_amountmin'].values

  model = DecisionTreeRegressor(max_depth=5)

  scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error')
  return np.mean(scores), np.std(scores)
```

```
feats = ['brand_cat']
run_model(feats)
```

```
df['manufacturernumber_cat'] = df['manufacturernumber'].factorize()[0]
```

```
feats = ['manufacturernumber_cat']
run_model(feats)
```

```
feats = ['brand_cat', 'manufacturernumber_cat']
run_model(feats)
```
